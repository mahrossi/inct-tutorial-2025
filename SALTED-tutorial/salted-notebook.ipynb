```python
""" Function to read xyz snapshots from a file and visualize them """

def read_xyz_snapshots(file_path):
    with open(file_path) as file:
        content = file.read()
    snapshots = content.strip().split("\n\n")  # Assuming each snapshot is separated by a blank line
    return snapshots


def read_xyz_snapshots_continuous(file_path):
    snapshots = []
    with open(file_path) as file:
        while True:
            header = file.readline()
            if not header:
                break  # End of file
            num_atoms = int(header.strip())
            comment = file.readline()  # Comment line, not used here but necessary to read

            # Read the atom lines for this snapshot
            atoms = [file.readline() for _ in range(num_atoms)]

            # Combine to form the snapshot
            snapshot = f"{num_atoms}\n{comment}{''.join(atoms)}"
            snapshots.append(snapshot)

    return snapshots


def visualize_snapshot(snapshots, index=0):
    view = py3Dmol.view(width=800, height=400)  # Adjust size as needed
    view.addModel(snapshots[index], "xyz")
    view.setStyle({"stick": {}})
    view.zoomTo()
    return view.show()
```

## Density fitting and its quality

In FHI-aims, electron density is represented using numerical atomic orbitals (NAOs):
$$
\rho (\vec{r}) = \sum_{I \mu J \nu} \rho_{I \mu J \nu} \phi_{I \mu} (\vec{r}) \phi_{J \nu}^* (\vec{r})
$$
where $\rho_{I \mu J \nu}$ are density matrix elements and $\phi_{I \mu} (\vec{r})$ are NAO basis functions, each consisting of a radial part and an angular part (spherical harmonics).

The **Resolution of Identity (RI)** method approximates electron density using auxiliary basis functions $Q_{Im}$.
Originally developed to accelerate Hartree-Fock exchange calculations, RI also provides an efficient density representation for machine learning models.

The RI expansion approximates density as:
$$
\begin{aligned}
\rho^{\mathrm{RI}}(\vec{r}) &= \sum_{Im} c_{Im} Q_{Im}(\vec{r}) \\
c_{Im} &= \sum_{Jn} (S^{-1})_{ImJn} \left\langle Q_{Jn} \middle| \rho \right\rangle
\end{aligned}
$$
The RI coefficients $c_{Im}$ serve as the learning targets for SALTED. Since auxiliary basis sets are typically only 3-10 times larger than the full density matrix representation, this dramatically reduces computational complexity and memory requirements ($N_\mathrm{NAO}^2 \gg 10 \times N_\mathrm{NAO} \sim N_\mathrm{RI}$).


We assess RI quality using the mean absolute error:
$$
\delta = \frac{
    \int \mathrm{d} \vec{r} \left| \rho^{\mathrm{SCF}}(\vec{r}) - \rho^{\mathrm{RI}}(\vec{r}) \right|
}{
    \int \mathrm{d} \vec{r} \rho^{\mathrm{SCF}}(\vec{r})
}
$$

This metric quantifies how accurately the auxiliary basis functions reproduce the true SCF electron density.

## Atomic environments and Visualization

The SOAP descriptor (or atomic environment) is atom-centered, and its corresponding kernel function (the one for Gaussian process regression) is invariant to translation and rotation in real space.

![SOAP descriptor](./SOAP_illustration_with_kernel.png)


The atomic density representation is
$$
\rho^{i,a} (\vec{r})
= \sum_j \delta_{a,a_j}
\exp \left( \dfrac{- |\vec{r} - \vec{r}_{ij}|^2}{2 \sigma_a^2} \right)
f_\mathrm{cut} (\vec{r}_{ij})
$$
This representation:
- Centers on atom $i$ and includes neighbors $j$ of species $a$
- Uses Gaussian smoothing with width $\sigma_a$ 
- Applies a cutoff function $f_\mathrm{cut}$ to limit the local environment


The invariant kernel function is
$$
k(M_i, M_j) = \sum_{aa'} \int_{\hat{R} \in \mathrm{SO_3}} \mathrm{d} \hat{R}
\left| \int \mathrm{d} \vec{r} \rho^{i,a}(\vec{r}) \rho^{j,a'}(\hat{R}\vec{r}) \right|^2
$$
which has been widely used in machine learning for learning molecular and materials properties.

The invariant kernel function is
$$
k(M_i, M_j) = \sum_{aa'} \int_{\hat{R} \in \mathrm{SO_3}} \mathrm{d} \hat{R}
\left| \int \mathrm{d} \vec{r} \rho^{i,a}(\vec{r}) \rho^{j,a'}(\hat{R}\vec{r}) \right|^2
$$
which has been widely used in machine learning for learning molecular and materials properties.

The atomic density is projected onto spherical harmonics to obtain rotationally covariant coefficients
$$
c_{nlm}^{i,a}
= \int \mathrm{d} \vec{r} R_n^*(r) Y_{l}^{m*}(\vec{r}) \rho^{i,a}(\vec{r})
$$
and the power spectrum (finally, we come to the data representation of descriptor / atomic environment) is
$$
p_{nn'l}^{i,aa'} = \frac{1}{\sqrt{2l+1}} \sum_m \left(c_{nlm}^{i,a}\right)^* c_{n'lm}^{i,a'}
$$

The final SOAP kernel combines power spectra across different atom types
$$
k (M_i, M_j) \propto \sum_{aa'nn'l} (p_{nn'l}^{i,aa'})^* (p_{nn'l}^{j,aa'})
$$
This kernel captures structural similarity between atomic environments.


The SALTED framework uses atomic environments for constructing kernel functions in Gaussian process regression. These atomic environments also serve as the basis for Subset of Regressors (SoR) selection, enabling efficient representation of the full training set through a carefully chosen subset of representative atomic configurations.

In this section, we will **compute atomic environments** for all structures in the training set, **perform sparsification** using Farthest Point Sampling (FPS) to select representative atomic environments for SoR, and **visualize the distribution** of atomic environments using dimensionality reduction techniques (t-SNE) to demonstrate the quality and coverage of the selected subset.

**Subset of Regressors (SoR)** approximates the full kernel matrix using only a representative subset $M$ of training points:
$$
K_{NN} \approx K_{NM} K_{MM}^{-1} K_{MN}
$$

This reduces the computational cost integrated with RKHS (see below). The subset $M$ is typically selected to capture the most important structural diversity in the training data.


**Farthest Point Sampling (FPS)** is a method used to select a subset of points from a larger set such that the selected points are as far apart from each other as possible.
This is particularly useful in scenarios where we want to ensure that the selected points represent the diversity of the original dataset.


In SALTED, the $\lambda=0$ descriptors form a vector space.
Starting from a random point, FPS iteratively selects the point that is farthest from the already selected points, ensuring the selected subset is well-distributed across the entire space of atomic environments and can effectively represent the full training set.

The selected subset of atomic environments serves as the **Subset of Regressors (SoR)** in the Gaussian process regression, reducing computational complexity while maintaining model accuracy through representative coverage of the chemical space.

## Computing the descriptors and kernels

Different from traditional SOAP descriptor, we use symmetry-adapted SOAP descriptor / kernels for tensorial properties prediction in SALTED.

The new symmetry-adapted kernel goes as
$$
\mathbf{k}^{\lambda} (M_i, M_j) = \int_{\hat{R} \in \mathrm{SO_3}} \mathrm{d} \hat{R} \mathbf{D}^{\lambda} (\hat{R})
\left| \int \mathrm{d} \vec{r} \rho^{i}(\vec{r}) \rho^{j}(\hat{R}^{-1} \vec{r}) \right|
$$

And the new symmetry-adapted power-spectrum can be calculated by the linear combination of the original SOAP coefficients:
$$
\mathbf{p}_{nn'll'}^{\lambda\nu} (M_i)
= (-1)^{\lambda-l'} \sum_m c_{nlm} (M_i) c_{n'l'(m-\mu)}^* (M_i)
\times (-1)^{m}
\begin{bmatrix}
\lambda & l' & l \\
\mu & (m-\mu) & -m
\end{bmatrix}
$$
and the kernel function is
$$
\mathbf{k}^{\lambda}_{\mu\nu} (M_i, M_j) \propto \sum_{nn'll'} \mathbf{p}_{nn'll'}^{\lambda\mu} (M_i) \mathbf{p}_{nn'll'}^{\lambda\nu} (M_j)
$$

As mentioned above, the **SoR** integrates with **Reproducing Kernel Hilbert Space (RKHS)** facilitates efficient Gaussian process regression in SALTED.

RKHS factorizes the kernel matrix using eigendecomposition:
$$
K_{MM} = V_{MD} \Lambda_{DD} V_{MD}^T
$$

This allows us to represent the kernel using feature vectors $\Psi$:
$$
\Psi_{ND} = K_{NM} V_{MD} \Lambda_{DD}^{-1/2}
$$
and the original kernel matrix $K_{NN}$ can be approximated as:
$$
K_{NN} \approx K_{NM} K_{MM}^{-1} K_{MN} \approx \Psi_{ND} \Psi_{ND}^T
$$

The SALTED prediction then becomes:
$$
c_N \approx \Psi_{ND} \tilde{w}_D
$$
where $\tilde{w}_D$ are the learned feature weights.
SoR + RKHS here dramatically reduces both memory requirements and computational complexity.

## Regression in two ways

SALTED minimizes spatial density MSE (equivalently a weighted L2 loss) between predicted and target RI coefficients:
$$
\mathcal{L}(\tilde{w}_D) = (\Psi_{ND} \tilde{w}_D - c_N^{\mathrm{DF}})^T S_{NN} (\Psi_{ND} \tilde{w}_D - c_N^{\mathrm{DF}}) + \eta \tilde{w}_D^T \tilde{w}_D
$$

The overlap matrix $S_{NN}$ weights the loss in the auxiliary basis space, while $\eta$ provides L2 regularization.


**Direct Inversion Method**

Taking the derivative and setting it to zero gives the normal equations
$$
\begin{aligned}
(\Psi_{ND}^T S_{NN} \Psi_{ND} + \eta \mathbb{I}_{DD}) \tilde{w}_D
& - \Psi_{ND}^T S_{NN} c_N^{\mathrm{DF}} = 0 \\
\Rightarrow (\mathbf{H} + \eta \mathbb{I}_{DD}) \tilde{w}_D
& - \vec{A} = 0
\end{aligned}
$$
where matrix $H = \Psi_{ND}^T S_{NN} \Psi_{ND}$ is the **Hessian matrix** (second derivative of the loss function), which is positive definite and symmetric.
The problem can be solved by directly inverting the Hessian, and the matrix size is $D \times D$.
This is feasible for small dimensions $D$ (e.g., $D < 10^{5}$).


**Conjugate Gradient (CG) Method** ([Wikipedia page](https://en.wikipedia.org/wiki/Conjugate_gradient_method))

For large feature dimensions $D$, directly inverting the Hessian becomes expensive. Instead, we can solve $H \tilde{w}_D = b$ iteratively using CG, which only requires matrix-vector products with $H$ rather than explicit inversion.

CG is particularly effective here because:
- The Hessian is positive definite (guaranteed convergence)
- We only need matrix-vector products, not the full matrix
- Convergence is typically fast for well-conditioned problems

## Validation by electron density

To evaluate the spatial electron density prediction error, we implement electron density RMSE:
$$
\Delta\rho
= \left(
    \frac{
        \sum_k \int \mathrm{d} \vec{r} \left| \rho_k^\mathrm{ML}(\vec{r}) - \rho_k^\mathrm{DF}(\vec{r}) \right|^2
    }{
        \sum_k \int \mathrm{d} \vec{r} \left| \rho_k^\mathrm{DF}(\vec{r}) - \rho_k^\mathrm{sph}(\vec{r}) \right|^2
    }
\right)^{1/2}
= \left(
    \frac{
        \sum_k \left( c_k^\mathrm{ML} - c_k^{\mathrm{DF}} \right)^\top S_{k} \left( c_k^\mathrm{ML} - c_k^{\mathrm{DF}} \right)
    }{
        \sum_k \left( c_k^\mathrm{DF} - c_k^{\mathrm{sph}} \right)^\top S_{k} \left( c_k^\mathrm{DF} - c_k^{\mathrm{sph}} \right)
    }
\right)^{1/2}
$$
Notice that the denominator is spatial electron density subtracted by the spherical average, so this electron density RMSE metric is actually very sensitive to spatial prediction error.
