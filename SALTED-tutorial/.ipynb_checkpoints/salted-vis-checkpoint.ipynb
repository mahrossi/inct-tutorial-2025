# Visualizing the SALTED tutorial

```python
import py3Dmol
import numpy as np
import matplotlib.pyplot as plt
import h5py
from ipywidgets import interact, IntSlider
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
import salted
import os
```

```python
# Function to read xyz snapshots from a file and visualize them
def read_xyz_snapshots(file_path):
    with open(file_path, 'r') as file:
        content = file.read()
    snapshots = content.strip().split('\n\n')  # Assuming each snapshot is separated by a blank line
    return snapshots
def read_xyz_snapshots_continuous(file_path):
    snapshots = []
    with open(file_path, 'r') as file:
        while True:
            header = file.readline()
            if not header:
                break  # End of file
            num_atoms = int(header.strip())
            comment = file.readline()  # Comment line, not used here but necessary to read
            
            # Read the atom lines for this snapshot
            atoms = [file.readline() for _ in range(num_atoms)]
            
            # Combine to form the snapshot
            snapshot = f"{num_atoms}\n{comment}{''.join(atoms)}"
            snapshots.append(snapshot)
            
    return snapshots
def visualize_snapshot(snapshots, index=0):
    view = py3Dmol.view(width=800, height=400)  # Adjust size as needed
    view.addModel(snapshots[index], 'xyz')
    view.setStyle({'stick': {}})
    view.zoomTo()
    return view.show()
```

# Visualize atomic environments and the ones chosen by FPS

```python
# initialize salted and compute atomic environments
! python3 -m salted.initialize
```

```python
file_path = 'equirepr_test/FEAT-0.h5'
# Read data
with h5py.File(file_path, 'r') as file:
    datasoap = np.array(file['descriptor'])

n_samples, n_atoms, n_features = datasoap.shape
datasoap_reshaped = datasoap.reshape(n_samples * n_atoms, n_features)
```

```python
# sparsify the descriptor selection based on the number Menv
# This sparsification is based on the (dense) features of angular momentum lambda = 0
! python3 -m salted.sparse_selection
```

```python
# Initialize t-SNE with desired parameters
tsne = TSNE(n_components=2, perplexity=30.0, max_iter=1000)

# Apply t-SNE to the reshaped data (using the same reshaping as for PCA)
data_tsne = tsne.fit_transform(datasoap_reshaped)

# get indexes
fps_indices = (np.loadtxt('equirepr_test/sparse_set_80.txt')).T[0]
fps_species = (np.loadtxt('equirepr_test/sparse_set_80.txt')).T[1]
selected_indices_h = [int(i) for i,j in zip(fps_indices, fps_species) if j==0]
selected_indices_o = [int(i) for i,j in zip(fps_indices, fps_species) if j==1]

# Plot chosen points along with the original t-SNE projection
plt.scatter(data_tsne[:, 0], data_tsne[:, 1], alpha=0.5, label='Full Data')
plt.scatter(data_tsne[selected_indices_h, 0], data_tsne[selected_indices_h, 1], color='red', edgecolor='k', label='FPS-Selected Points H', s=100, linewidths=2)
plt.scatter(data_tsne[selected_indices_o, 0], data_tsne[selected_indices_o, 1], color='yellow', edgecolor='k', label='FPS-Selected Points O', s=100, linewidths=2)


# Add labels and legend
plt.title('t-SNE Projection')
plt.xlabel('Dimension 1')
plt.ylabel('Dimension 2')
plt.legend()

plt.show()
```

# Regression 

We will do the regression in two ways: by direct inversion and by minimization of loss. The latter algorithm is the most suitable for large systems or large datasets

```python
# inversion first. make the (hessian) matrix to be inverted
# We will be using 70 structure to train - Ntrain. The algorithm reports their indexes
! python3 -m salted.hessian_matrix
```

```python
# carry out matrix inversion
! python3 -m salted.solve_regression
```

```python
#TODO: Think about visualization of regression weights, stored in regrdir_test/M80_zeta2.0/weights_N70_reg-6.npy
# explain the dimension of this array: total dimension of Psi (RKHS vector): a*n*lambda*N*D = 7117.
# can we separate the plot into the different dimensions and lambda channels?
weights = np.load('regrdir_test/M80_zeta2.0/weights_N70_reg-6.npy')
plt.plot(np.arange(len(weights)), weights, marker='o', linestyle='-')
plt.xlabel('Index')
plt.ylabel('Weight')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Now alternative regression by minimizing the loss function with conjugate gradient. We copy previous weights just to plot later
! cp regrdir_test/M80_zeta2.0/weights_N70_reg-6.npy regrdir_test/M80_zeta2.0/weights_N70_reg-6-INV.npy
! python3 -m salted.minimize_loss
```
